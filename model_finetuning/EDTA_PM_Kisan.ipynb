```python
import pandas as pd

# File path to the CSV file
file_path = "/Users/chaitanyakartik/Projects/Newron_GR/data/XL_Sheets/Actual_Griviences/Cleaned_PM_Kisan.csv"

# Load the CSV file into a pandas DataFrame
df = pd.read_csv(file_path)

# Create the department string for each row, excluding NaN in the last field
df['Department_String'] = df.apply(
    lambda row: f"{row['Department']}/{row['Line Department']}/{row['Service Name']}/{row['Grievance Name']}/{row['Grievance SubCategory1']}" +
                (f"/{row['Grievance SubCategory2']}" if pd.notna(row['Grievance SubCategory2']) else ""),
    axis=1
)

# Create a dictionary with G.Id as the key and Department_String as the value
gid_department_dict = dict(zip(df['G.Id'], df['Department_String']))

# Print the dictionary to verify
print(gid_department_dict)
```

```python
gid_department_dict
```

```python
# Initialize counters
in_count = 0
not_in_count = 0

# Initialize a dictionary to count occurrences for each category
category_counts = {category: 0 for category in categories}

# Loop through the dictionary values
for gid, department_string in gid_department_dict.items():
    if department_string.upper() in categories:
        in_count += 1
        category_counts[department_string.upper()] += 1
    else:
        not_in_count += 1

# Display results
print(f"Total matches found: {in_count}")
print(f"Total not found: {not_in_count}")
print("\nCategory-wise counts:")
for category, count in category_counts.items():
    print(f"Count: {count}   Category:  {category}")
```

### Now we need to clean up our data to create train and test data
- translate all griviences from kannada to english
- create a train test split for the data and store as csv files
- create a jsonl file for this


```python
import pandas as pd
from ..utils.chat_utils import translate_query

# File path to the CSV file
file_path = "/Users/chaitanyakartik/Projects/Newron_GR/data/XL_Sheets/Actual_Griviences/Cleaned_PM_Kisan.csv"

# Load the CSV file into a pandas DataFrame
df = pd.read_csv(file_path)

# Create the department string for each row, excluding NaN in the last field
df['target_grievance_category'] = df.apply(
    lambda row: f"{row['Department']}/{row['Line Department']}/{row['Service Name']}/{row['Grievance Name']}/{row['Grievance SubCategory1']}" +
                (f"/{row['Grievance SubCategory2']}" if pd.notna(row['Grievance SubCategory2']) else ""),
    axis=1
)

# Translate the grievance descriptions
df['translated_grievance'] = df['Grievance Description'].apply(translate_query)

# Select the required columns
output_df = df[['G.Id', 'translated_grievance', 'target_grievance_category']]

# Rename columns to match the required format
output_df.rename(columns={'G.Id': 'G_ID'}, inplace=True)

# Save to a new CSV file
output_file_path = "/Users/chaitanyakartik/Projects/Newron_GR/data/translated_grievances.csv"
output_df.to_csv(output_file_path, index=False)

print(f"CSV file created at: {output_file_path}")
```

```python
import pandas as pd
import os

# File path to the input CSV file
input_file_path = "/Users/chaitanyakartik/Projects/Newron_GR/data/Fine_Tuning_data/PM_Kisan/PM_Kisan_Grivience_data.csv"

# Load the CSV file into a pandas DataFrame
df = pd.read_csv(input_file_path)

# Perform an 80-20 train-test split
train_df = df.sample(frac=0.8, random_state=42)  # Randomly select 80% of the data for training
test_df = df.drop(train_df.index)  # The remaining 20% for testing

# Define output file paths
output_dir = os.path.dirname(input_file_path)
train_file_path = os.path.join(output_dir, "train.csv")
test_file_path = os.path.join(output_dir, "test.csv")

# Save the train and test datasets to CSV files
train_df.to_csv(train_file_path, index=False)
test_df.to_csv(test_file_path, index=False)

print(f"Train data saved to: {train_file_path}")
print(f"Test data saved to: {test_file_path}")
```
