```python
import pandas as pd
import json
import numpy as np

def build_tree(df, level=0):
    levels = [
        ('department_name', 'categ_id'),
        ('category_description', 'categ_id'),
        ('sub_categ1_english_desc', 'sub_categ1_id'),
        ('sub_categ2_english_desc', 'sub_categ2_id'),
        ('sub_categ3_english_desc', 'sub_categ3_id'),
        ('sub_categ4_english_desc', 'sub_categ4_id'),
    ]
    if level >= len(levels):
        return []

    name_col, id_col = levels[level]
    nodes = []
    for name, group in df.groupby(name_col):
        if pd.isna(name):
            continue
        categ_id = str(group[id_col].iloc[0])
        if pd.isna(categ_id):
            continue
        children = build_tree(group, level + 1)
        node = {
            "name": name,
            "categ_id": categ_id,
            "level": str(level),
        }
        if children:
            node["children"] = children
        nodes.append(node)
    return nodes

tree = build_tree(master_df)
print(tree)
```

```python
with open('departments_tree.json', 'w', encoding='utf-8') as f:
    json.dump(tree, f, ensure_ascii=False, indent=2)
```
